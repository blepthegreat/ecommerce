## Exploratory Data Analysis (EDA)

### Explore ...

```python
# Merge order_items_dataset_df with products_dataset_df to get product_category_name
merged_df = pd.merge(order_items_dataset_df, products_dataset_df[['product_id', 'product_category_name']], on='product_id', how='left')

# Merge the result with translation_df to get product_category_name_english
merged_df = pd.merge(merged_df, translation_df[['product_category_name', 'product_category_name_english']], on='product_category_name', how='left')

# Now merged_df contains both product_category_name and product_category_name_english

# Menghitung jumlah transaksi berdasarkan kategori produk dan metode pembayaran
# Note: 'payment_type' from order_payment_df is used as a proxy for 'payment_method'
payment_category_counts = merged_df.groupby(['product_category_name_english', order_payment_df['payment_type']]).size().unstack(fill_value=0)

# Rename columns for clarity and to match the barplot arguments
payment_category_counts = payment_category_counts.rename_axis(index='product_category_name_english') # Rename index
payment_category_counts = payment_category_counts.reset_index() # Reset index to column

# Melt the DataFrame to long format for easier plotting with seaborn
payment_category_counts = payment_category_counts.melt(id_vars=['product_category_name_english'], var_name='payment_method', value_name='transaction_count')

# Menentukan Top 10 kategori produk berdasarkan jumlah transaksi total
top_10_categories = payment_category_counts.groupby('product_category_name_english')['transaction_count'].sum().sort_values(ascending=False).head(10).index

# Filter untuk hanya menampilkan Top 10 kategori produk
top_10_payment_category_counts = payment_category_counts[payment_category_counts['product_category_name_english'].isin(top_10_categories)]

# Visualisasi
plt.figure(figsize=(14, 8))
sns.barplot(data=top_10_payment_category_counts,
            x='product_category_name_english',
            y='transaction_count',
            hue='payment_method',
            palette='Set2')

plt.title("Top 10 Produk Kategori Terlaris Berdasarkan Metode Pembayaran", fontsize=16)
plt.xlabel("Kategori Produk", fontsize=12)
plt.ylabel("Jumlah Transaksi", fontsize=12)
plt.xticks(rotation=45)
plt.legend(title='Metode Pembayaran', loc='upper right')
plt.tight_layout()
plt.show()
```

Berdasarkan grafik "Top 10 Produk Kategori Terlaris Berdasarkan Metode Pembayaran" yang ditampilkan, berikut adalah beberapa insight penting:

1. Kartu kredit (credit_card) adalah metode pembayaran yang paling dominan di semua kategori produk, dengan jumlah transaksi jauh lebih tinggi dibandingkan metode pembayaran lainnya.

2. Kategori "bed_bath_table" memiliki jumlah transaksi tertinggi dengan kartu kredit, mencapai sekitar 7.500 transaksi, diikuti oleh "health_beauty" (sekitar 6.500) dan "sports_leisure" (sekitar 5.800).

4. Kartu debit (debit_card) dan voucher adalah metode pembayaran yang paling jarang digunakan, dengan jumlah transaksi yang relatif rendah di semua kategori.

5. Kategori "auto" memiliki jumlah transaksi kartu kredit paling rendah dibandingkan kategori lainnya, tetapi tetap menjadi metode pembayaran dominan untuk kategori tersebut.

6. Voucher sebagai metode pembayaran menunjukkan penggunaan yang relatif konsisten di semua kategori, meskipun jumlahnya kecil.

## Analisis Lanjutan (Opsional)

```python
import pandas as pd
import numpy as np

# Load dataset
folder_path = "/content/drive/My Drive/E-Commerce Public Dataset/"
order_payment_df = pd.read_csv(folder_path + "order_payments_dataset.csv")
orders_df = pd.read_csv(folder_path + "orders_dataset.csv")

# Konversi tanggal transaksi ke format datetime
orders_df["order_purchase_timestamp"] = pd.to_datetime(orders_df["order_purchase_timestamp"])

# Gabungkan dataset pembayaran dengan informasi tanggal transaksi
order_payment_df = order_payment_df.merge(
    orders_df[["order_id", "order_purchase_timestamp"]],
    on="order_id",
    how="left"
)

# Menentukan tanggal referensi (hari terakhir dalam dataset)
latest_date = order_payment_df["order_purchase_timestamp"].max()

# Hitung RFM berdasarkan metode pembayaran
rfm_payment = order_payment_df.groupby("payment_type").agg({
    "order_purchase_timestamp": lambda x: (latest_date - x.max()).days,  # Recency (R)
    "order_id": "count",  # Frequency (F)
    "payment_value": "sum"  # Monetary (M)
}).reset_index()

# Ganti nama kolom agar lebih mudah dibaca
rfm_payment.columns = ["payment_type", "Recency", "Frequency", "Monetary"]

# Tampilkan hasil
print(rfm_payment.sort_values(by="Frequency", ascending=False))

import matplotlib.pyplot as plt
import seaborn as sns

# Urutkan berdasarkan Frequency agar lebih mudah dibaca
rfm_payment_sorted = rfm_payment.sort_values(by="Frequency", ascending=False)

# 🔹 Visualisasi Frequency, Monetary, dan Recency dengan Bar Chart
fig, axes = plt.subplots(3, 1, figsize=(10, 15))

# Frequency (Jumlah Transaksi per Metode Pembayaran)
sns.barplot(x=rfm_payment_sorted["Frequency"], y=rfm_payment_sorted["payment_type"], palette="Blues", ax=axes[0])
axes[0].set_title("🔹 Jumlah Transaksi per Metode Pembayaran", fontsize=14)
axes[0].set_xlabel("Frequency (Jumlah Transaksi)")
axes[0].set_ylabel("Metode Pembayaran")

# Monetary (Total Nilai Transaksi per Metode Pembayaran)
sns.barplot(x=rfm_payment_sorted["Monetary"], y=rfm_payment_sorted["payment_type"], palette="Greens", ax=axes[1])
axes[1].set_title("💰 Total Nilai Transaksi per Metode Pembayaran", fontsize=14)
axes[1].set_xlabel("Monetary (Total Nilai Transaksi)")
axes[1].set_ylabel("Metode Pembayaran")

# Recency (Seberapa Baru Transaksi Terakhir)
sns.barplot(x=rfm_payment_sorted["Recency"], y=rfm_payment_sorted["payment_type"], palette="Reds", ax=axes[2])
axes[2].set_title("🕒 Recency (Hari Sejak Transaksi Terakhir)", fontsize=14)
axes[2].set_xlabel("Recency (Hari Sejak Transaksi Terakhir)")
axes[2].set_ylabel("Metode Pembayaran")

plt.tight_layout()
plt.show()

# 🔹 Heatmap untuk Melihat Pola RFM
plt.figure(figsize=(10, 6))
sns.heatmap(rfm_payment_sorted.set_index("payment_type"), annot=True, fmt=".0f", cmap="coolwarm", linewidths=0.5)
plt.title("🔥 Heatmap RFM Metode Pembayaran", fontsize=14)
plt.show()
```
